```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer# used to convert text data to numerical data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data collection and preprocessing
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
#replace null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the 1st five rows
mail_data.head()
```

```python
#checking the number of rows and columns in the data frame
mail_data.shape
```

```python
#lavel encoding ,ham->1,spam->0
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
#separating the data as texts and level
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(Y)
```

```python
# SPLITTING THE DATA INTO TRAIN AND TEST
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
  #logistic regression model training with training data
  model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
from os import pread
#prediction on training data
preaciction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, preaciction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
#prediction is test data
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
#building a predictive system
input_mail = ["XXXMobileMovieClub: To use your credit, click the WAP link in the next txt message or click here>> http://wap. xxxmobilemovieclub.com?n=QJKGIGHJJGCBL"]
```

```python
#convert text to features
input_data_features = feature_extraction.transform(input_mail)
#making prediction
prediction = model.predict(input_data_features)
print(prediction)
if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
